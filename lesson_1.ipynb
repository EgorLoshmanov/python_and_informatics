*Компьютерная программа* — список команд (инструкций) для компьютера.

*Язык программирования* — набор определённых правил, по которым компьютер может понимать команды (инструкции) и выполнять их.

## Переменные. Типы данных

*Переменная* — это имя, по которому хранится значение в программе.

*Тип данных* — это свойство переменной, которое определяет какие значения она может хранить и какие операции над ней допустимы.

````python
number1 = 10  # Имя переменной = Значение переменной 
number2 = 1000

print(number1, number2)  # Вывод информации на экран
````

````python
# Переменные могут хранить различные данные: число, текст и т.д
number = 36  # Число
line = "hello, world!"  # Текст. Обязательно выделяется кавычками!!!
line_number = '36'  # Также текст

print(number)
print(line)
print(line_number)
````

````python
# Переменная хранит последнее переданное ей значение
x = 10
x = x + 1000  # Сейчас в переменной x лежит число 1010

x = 0  # Присваиваем значение 0
print(x)
````

````python
# В ходе выполнения программы, тип данных в переменной может изменяться
variable = 10  # int - целое число
print(variable, type(variable))  # type(<переменная>) - выводит тип переменной

variable = "Moscow"
print(variable, type(variable))  # str - строка с текстом
````

## Ввод-вывод данных

Для вывода данных на экран используется команда `print()`.

Внутри круглых скобок пишем, что хотим вывести на экран. Если это текст, то обязательно указываем его внутри кавычек. Кавычки могут быть одинарными (') или двойными ("). До и после текста мы ставим только одинаковые кавычки.

````python
print('Я изучаю язык Python')  # Одинарные кавычки
print("Я изучаю язык Python")  # Двойные
````

````python
# Разные кавычки
print('Я изучаю язык Python")  
````

### Аргументы команды print()
То, что мы пишем в круглых скобках у команды print(), называется **аргументами** команды.

*Аргументы* – это конкретные значения, которые вы передаете функции при ее вызове.

Команда print() позволяет указывать несколько аргументов, в таком случае их надо отделять запятыми.


````python
print('Я', 'учусь', 'программировать', 'на', 'Python!')
````

Обратите внимание, в качестве разделителя при выводе данных между аргументами команды используется символ пробела. По умолчанию команда print() добавляет ровно один пробел между всеми своими аргументами.

````python
print('1', '2', '4', '8', '16')
````

## Важные замечания

### Замечание 1
Команда print() записывается только маленькими буквами, другое написание недопустимо, так как в Python строчные и заглавные буквы различны.


### Замечание 2
Каждая последующая команда print() выводит указанный текст с новой строки.


````python
print("Какой-то текст")
print("Всё также какой-то текст")
````

### Замечание 3
Команда print() с пустым списком аргументов вставляет пустую строку.


````python
print("Какой-то текст")
print()
print("Всё также какой-то текст")
````

### Замечание 4
Если в тексте есть кавычки, то снаружи нужно поставить кавычки другого вида:
если внутри одинарные — снаружи двойные, и наоборот.



````python
print("Company 'Apple'")
print('Company "Apple"')
````

## Резюме

*Переменная* – ячейка (сундук), куда сохраняются данные. Любая переменная имеет имя и значение.

*Значение переменной* — это сохранённая в ней информация: текст, число и т.д.

*Тип данных* — это вид информации, который может храниться в переменной: число, текст и т.д. Тип определяет, какие операции можно выполнять с этим значением.

Команда `print()` используется для вывода данных на экран. В круглых скобках указываются данные, которые нужно вывести (аргументы). Если это текст, его нужно заключить в одинарные или двойные кавычки. Команда `print()` позволяет указывать несколько аргументов, в таком случае их надо отделять запятыми.

*Аргументы* – это конкретные значения, которые передаются функции при ее вызове
