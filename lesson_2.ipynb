# Урок 2. Ввод данных. Целочисленная арифметика

Раньше программы выводили только тот текст, который был *заранее* записан в коде. Но программы могут работать и с данными, которые становятся известны только во время работы. То есть программа может что-то считать (например, с клавиатуры) и потом это использовать.

Для считывания данных в языке Python используется команда `input()`.



Рассмотрим программу
```
print('Как тебя зовут?')     # вывод текста

print('Привет,', input())    # ввод текста и вывод текста
```

Сначала программа печатает текст на экран 'Как тебя зовут?'. Далее программа будет ждать от пользователя ввода данных. Ввод данных реализуется с помощью команды `input()`. И в конце произойдет вывод текста вместе с введенными от пользователя данными.

Команда `input()` всегда пишется с круглыми скобками. Она работает так: когда программа доходит до места, где есть `input()`, она ждет, пока пользователь введет текст с клавиатуры (ввод завершается нажатием клавиши **Enter**). Введенная строка подставляется на место `input()`

````python
print('Как тебя зовут?')     # вывод текста

print('Привет,', input())    # ввод текста и вывод текста
````

Здесь, когда программа попросила ввести данные (текст) с клавиатуры, мы написали слово "Андрей" и это слово подставилось вместо `input()`

## Напоминание
Команда `print()` выводит данные на экран.

Команда `input()` считывает данные, введенные с клавиатуры.

## Важные замечания

### Замечание 1
Python ждет, что пользователь что-то введет с клавиатуры ровно столько раз, сколько команд `input()` встречается в программе. Каждый `input()` завершается нажатием **Enter** на клавиатуре.



````python
# Пример

print("Как тебя зовут?")
first_name = input()
print("Какая у тебя фамилия?")
last_name = input()

print("Привет,", first_name, last_name)
````

В этой программе у нас две команды `input()`. В первый раз программа спрашивает имя человека, а во второй - фамилию. Имя кладётся в переменную `first_name`, а фамилия в `last_name`. В последней строке вместо `first_name` и `last_name` подставляются значения переменных, то есть `Егор` и `Лошманов`

### Замечание 2
Очень часто перед считыванием данных мы печатаем некоторый текст, чтобы пользователь, который вводит эти данные, понимал, что именно от него требуется. Например, в программе:
```
print('Как тебя зовут?')
name = input()
```
Мы сначала выведем текст "Как тебя зовут?", а уже потом считаем данные.

Поскольку так делают достаточно часто, то в языке Python можно выводить текст, передавая его в качестве аргумента в команду `input()`. Предыдущий код можно переписать так:
```
# сначала тут печатается строка 'Как тебя зовут', а потом принимается на вход имя
name = input('Как тебя зовут?')
```
То есть команда `input()` при наличии аргументов внутри нее отрабатывает одновременно как вывод текста, а потом ввод текста (именно в этом порядке).


### Замечание 3

Мы можем принимать сразу несколько строк, а потом со всеми ними работать.

````python
first = input()
second = input()

print('Я', first, 'и', second)
````

### Замечание 4
Переменные можно вводить в любой момент (не только в самом начале программы):

````python
name_1 = input()                          # принимаем имя первого человека
print('Привет,', name_1, '!')             # делаем первый вывод

name_2 = input()                          # принимаем имя второго человека
print('Здравствуйте,', name_2, '.')       # делаем второй вывод
````
